```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
authors = pd.read_csv('authors_all.csv', encoding = 'utf-8', sep = ";")
```

```python
#subset authors born later than 1830 or null
authors2 = authors[(authors['Birth year'] >1830) |(authors['Birth year'].isnull())]
```

```python
#subset of authors who died earlier than 1900 or null
authors3 = authors2[(authors2['Death year']>1900)|(authors2['Death year'].isnull())]
```

```python
#create new column for later calculations
authors3.insert(loc = authors3.shape[1], column ='Year_Latest_Artwork', value = np.nan)
```

```python
for i in range(authors.shape[0]):
    if (i in authors3.index):
        if(authors3.loc[i].artworks_creation_years is not np.nan):
            year = max(authors3.loc[i].artworks_creation_years.split('|'))
            authors3.at[i, 'Year_Latest_Artwork']= year
```

```python
#subset authors born later than 1830 or null
authors4 = authors3[(authors3['Year_Latest_Artwork'] > 1900) |(authors3['Year_Latest_Artwork'].isnull())]
```

```python
#write to csv file
authors4.to_csv('filtered_authors.csv', sep=',', encoding = 'utf-8-sig')
```
