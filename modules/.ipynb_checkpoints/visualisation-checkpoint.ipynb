```python
import re
import pandas as pd
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
```

```python
ex = pd.read_csv('../data/ALL_EXHIBITIONS.csv', sep=';', encoding='utf-8', low_memory=False)
art = pd.read_csv('../data/ALL_ARTWORKS.csv', sep=',', encoding='utf-8', low_memory=False)
```

```python
art['collection'].dropna().unique()
```

```python
G = nx.Graph()
exhib_matrix = {}
for ind, e in ex.iterrows():
    if ind % 1000 == 0:
        print("%i / %i, graphe : %i" % (_, len(ex), len(G.nodes)))
    e_list = re.split(r'\|', e['artworks'])
    if len(e_list) > 1:
        G.add_node(str(ind))
        exhib_matrix[ind] = e_list
        for a in e_list:
            aa = art[art["_id"] == int(a)]["collection"].tolist()[0]
            for other_exhib in exhib_matrix:
                if a in exhib_matrix[other_exhib]:
                    G.add_edge(str(ind), str(other_exhib), weight = 1)    
```

```python
colors = plt.get_cmap("tab10")
pos = nx.spring_layout(G)
fig, ax = plt.subplots(figsize=(20,20))
color = "blue"
nodes = nx.draw_networkx_nodes(G, pos, nodelist=G.nodes, node_color=color, node_size=30)
nodes.set_edgecolor('black')
weights = [G[u][v]['weight'] for u, v in G.edges]
nx.draw_networkx_edges(G, pos, width=weights)
plt.axis('off')
plt.show()
```
